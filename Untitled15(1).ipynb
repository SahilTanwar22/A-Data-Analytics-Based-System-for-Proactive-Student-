```python
# ======================================
# STEP 2: Upload Dataset
# ======================================
from google.colab import files
import pandas as pd

uploaded = files.upload()  # Upload "Activity_tracking_sheet 3.csv"

# Load with proper encoding
df = pd.read_csv("/content/sample_data/vle.csv", encoding="latin1")

print("✅ Dataset Shape:", df.shape)
df.head()
```

```python
# ======================================
# STEP 3: Preprocess Data
# ======================================
# Drop identifiers (not useful for prediction)
id_cols = ["Child Name", "Email Address", "Contact"]
df_clean = df.drop(columns=[col for col in id_cols if col in df.columns])

# Convert activity columns into binary (1 = completed/pass, 0 = not completed)
def to_binary(x):
    if isinstance(x, str):
        if "Completed" in x:   # includes "Completed" and "Completed (achieved pass grade)"
            return 1
        elif "Not completed" in x:
            return 0
    return 0  # default for NaN or other cases

df_binary = df_clean.applymap(to_binary)

# Add completion_rate (% of completed activities per student)
df_binary["completion_rate"] = df_binary.mean(axis=1)

# Create target: at-risk if <70% completion
df_binary["target"] = df_binary["completion_rate"].apply(lambda x: 1 if x < 0.7 else 0)

print("✅ Target distribution:\n", df_binary["target"].value_counts())
df_binary.head()
```

```python
# ======================================
# STEP 4: Define Features and Target
# ======================================
X = df_binary.drop("target", axis=1)
y = df_binary["target"]

print("Feature matrix shape:", X.shape)
```

```python
# ======================================
# STEP 5: Train Random Forest and XGBoost
# ======================================
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import numpy as np

rf = RandomForestClassifier(n_estimators=100, random_state=42)
xgb = XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

rf_scores = cross_val_score(rf, X, y, cv=skf, scoring="f1")
xgb_scores = cross_val_score(xgb, X, y, cv=skf, scoring="f1")

print("Random Forest F1 (CV):", np.mean(rf_scores))
print("XGBoost F1 (CV):", np.mean(xgb_scores))
```

```python
# ======================================
# STEP 6: Train and Evaluate XGBoost
# ======================================
xgb.fit(X, y)

from sklearn.metrics import classification_report, roc_auc_score

y_pred = xgb.predict(X)
print("\nXGBoost Classification Report:\n")
print(classification_report(y, y_pred))
print("XGBoost AUC:", roc_auc_score(y, y_pred))
```

```python
# ======================================
# STEP 7: Deep Neural Network
# ======================================
import tensorflow as tf
from tensorflow import keras

X_np = X.values
y_np = y.values

model = keras.Sequential([
    keras.layers.Dense(128, activation="relu", input_shape=(X_np.shape[1],)),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

history = model.fit(X_np, y_np, epochs=20, batch_size=16, validation_split=0.2, verbose=1)
```

```python
# ======================================
# STEP 8: Plot Training Accuracy
# ======================================
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# ======================================
# STEP 9: SHAP Explainability
# ======================================
import shap

explainer = shap.TreeExplainer(xgb)
shap_values = explainer.shap_values(X)

shap.summary_plot(shap_values, X)
```
